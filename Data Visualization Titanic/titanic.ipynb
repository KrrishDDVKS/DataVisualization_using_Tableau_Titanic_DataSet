```python
df.drop_duplicates()
dft.drop_duplicates()
df['Cabin'].values
dft['Cabin'].values
df1 = df['Cabin'].str.split(expand=True) # differentiating the cabins more than one taken.
dft1 = dft['Cabin'].str.split(expand=True) # differentiating the cabins more than one taken.
df1.columns = ['c{}'.format(x+1) for x in df1.columns]
dft1.columns = ['c{}'.format(x+1) for x in dft1.columns]
df1.info()
dft1.info()
df=df.join(df1)
dft=dft.join(dft1)
```

```python
df.drop(columns='c4') #Removed unwanted extra differentiated cabin.
dft.drop(columns='c4') #Removed unwanted extra differentiated cabin.
```

```python
for i in range(1,4):
    df[f'C{i}']=df[f"c{i}"].str[0]
    dft[f'C{i}']=dft[f"c{i}"].str[0]
    df[f'c_{i}']=df[f"c{i}"].str[1:] # The differentiated cabins are again divided as Class and number
    dft[f'c_{i}']=dft[f"c{i}"].str[1:] # The differentiated cabins are again divided as Class and number
df.head()
```

```python
dft.head(10)
```

```python
df['c_1']=df['c_1'].replace('',np.nan) 
dft['c_1']=dft['c_1'].replace('',np.nan) 
```

```python
df=df.astype({'c_1':float,'c_2':float, 'c_3':float})
dft=dft.astype({'c_1':float,'c_2':float, 'c_3':float})
```

```python
for i in range(1,4):
    df[f'c_{i}']=df[f'c_{i}'].fillna(int(df[f'c_{i}'].mean()))
    dft[f'c_{i}']=dft[f'c_{i}'].fillna(int(dft[f'c_{i}'].mean()))
    df[f'c_{i}']=df[f'c_{i}'].astype('int')
    dft[f'c_{i}']=dft[f'c_{i}'].astype('int')
```

```python
df=df.drop(['c1','c2','c3','c4','Cabin','Ticket','Name','PassengerId'],axis=1)
xtt=dft
dft=dft.drop(['c1','c2','c3','c4','Cabin','Ticket','Name','PassengerId'],axis=1)
```

```python
en=encoder()                                     #Converted all categorical values into numerical.
dft['Sex']=en.fit_transform(dft['Sex'])

dft['Embarked']=en.fit_transform(dft['Embarked'])
df['C1']=en.fit_transform(df['C1'])
dft['C1']=en.fit_transform(dft['C1'])
df['C2']=en.fit_transform(df['C2'])
dft['C2']=en.fit_transform(dft['C2'])
df['C3']=en.fit_transform(df['C3'])
dft['C3']=en.fit_transform(dft['C3'])
```

```python
df['C1'].replace(df['C1'].max(),int(df['C1'].mean()),inplace=True) # filled the missing values
dft['C1'].replace(dft['C1'].max(),int(dft['C1'].mean()),inplace=True) # filled the missing values
df['C2'].replace(df['C2'].max(),int(df['C2'].mean()),inplace=True)
dft['C2'].replace(dft['C2'].max(),int(dft['C2'].mean()),inplace=True)
df['C3'].replace(df['C3'].max(),int(df['C3'].mean()),inplace=True)
dft['C3'].replace(dft['C3'].max(),int(dft['C3'].mean()),inplace=True)
dft['Embarked'].replace(dft['Embarked'].max(),int(dft['Embarked'].mean()),inplace=True)
df['Age'].fillna(int(df['Age'].mean()),inplace=True)
dft['Age'].fillna(int(dft['Age'].mean()),inplace=True)
```
